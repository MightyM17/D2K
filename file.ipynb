```python
import re
import requests
from bs4 import BeautifulSoup
import random
import pandas as pd

def scrape(url):
    _url = requests.get(url=url,)

    soup = BeautifulSoup(_url.content, 'html.parser')
    
    quotes=[]
    authors=[]
    
    allLinks = soup.find(id="hs_cos_wrapper_post_body").find_all("h3")
    for link in allLinks:
        pattern = r'[0-9]'
        authors.append(re.sub(pattern, '', (str)(link.text)).replace(". ",""))
    
    allLinks = soup.find(id="hs_cos_wrapper_post_body").find_all("p")
    for link in allLinks:
        temp = (str)(link.text)
        if(temp.find("“") != -1):
            temp = temp.replace("“", "")
            temp = temp.replace("”", "")
            quotes.append(temp)

    lent = min(len(authors), len(quotes))
    print(quotes)
    df = pd.DataFrame(quotes[:lent], authors[:lent])
    df.to_csv("a.csv", header=False, mode='a',)

scrape('https://www.scuba.io/blog/48-analytics-quotes-experts')
```
